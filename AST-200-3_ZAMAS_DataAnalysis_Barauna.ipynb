```python
import matplotlib.pyplot as plt
import numpy as np

pontos = ['pontos1','pontos2','pontos3'] # criando uma lista com no nome das lista 
entrada = ['Codigo_Fonte/pontos1.dat','Codigo_Fonte/pontos2.dat','Codigo_Fonte/pontos3.dat'] #criando umas lista com endereço dos .dat
for f in range (0,len(pontos)):
    pontos[f] = []  #abre uma lista em branco
    with open(entrada[f]) as data: #abre o arquivo e chama dele de data
        for line in data: #loop para separação dos dados que estão separados por um 'espaco'
            pontos[f].append([(x) for x in line.split()])
for f in range (0,len(pontos)):
    for i in range(1,len(pontos[f])):  #para todos as linhas dentro da lista 'pontos' (menos a primeira)
        for j in range(0,len(pontos[f][0])-1):#para todos os valores dentro das COLUNAS 
            pontos[f][i][j] = pontos[f][i][j].replace('D','e')#troque todos os 'D' por 'e'        
```

## Questão 10

Gradiente de Temperatura Radiativa
\begin{equation}
    \frac{dT}{dr} = \frac{3k_R\rho (r)L(r)}{16\pi acr^2 T(r)^3}.
\end{equation}

Gradiente de Temperatura Adiabática
$$ \frac{dT}{dr} = \frac{\Gamma_2-1}{\Gamma_2}\frac{T(r)}{P(r)} \frac{dP(r)}{dr} $$

$$\frac{dT}{dr} = \frac{\Gamma_2-1}{\Gamma_2}\frac{T(r)}{P(r)} \frac{GM(r)\rho(r)}{r^2}$$

```python
#Constantes
a = 7.5658*10**-15 #𝑒𝑟𝑔 𝑐𝑚−3 𝐾−4
c = 3*10**10 # 𝑐𝑚/s-1
L_sol = 3.83*10**33 # ergs/segundos# Luminosidade de Sol
R_sol = 6.96*10**10	# centimetros  # Raio do Sol
gamma = 5/2
G = 6.67*10**(-8)
```

```python
dTdr1 = []
dTdr2 = []
dTdr3 = []
diferenca_dos_gradientes = []
for i in range (0,conv):
    dTdr1.append( 3*kaptot[i]*densidade[i]*Lumin[i]*L_sol/(16*np.pi*a*c*(raio1[i]*R_sol)**2*Temperatura[i]**3))

for i in range (0,conv):
    dTdr2.append( (gamma-1)/gamma * Temperatura[i]/pressao_t[i] * G*massa_M[i]*densidade[i]/(raio1[i]**2))

for i in range (0,conv):
    dTdr3.append(dTdr1[i] - dTdr2[i])

plt.plot(raio1[:conv],np.log10(dTdr1[:conv]),color='blue',label='Gradiente de temperaura radiativo')
plt.plot(raio1[:conv],np.log10(dTdr2[:conv]),color='red',label='Gradiente de temperaura adiabatico')
plt.axvline(x=r_convergencia,color='red',label='Ponto de Convergência',linestyle='--')
plt.xlabel('Raio ($R_\odot$)')
plt.ylabel('$\log(dT/dr)$ $(K cm^{-1})$')
plt.title('Gradiente de Temperaura')
plt.legend()
plt.xlim(0.76,1)
plt.savefig('graficos/grafico_i.pdf')
plt.show()
```
